## Section 3.1
How many seconds are in an hour?

```python
60 * 60
```

## Section 3.2
Assign seconds per hour to a variable.

```python
seconds_per_hour = 60 * 60
seconds_per_hour
```

## Section 3.3
How many seconds are in a day?

```python
seconds_per_hour * 24
```

## Section 3.4
Assign seconds per day to a variable.

```python
seconds_per_day = seconds_per_hour * 24
seconds_per_day
```

## Section 3.5
Divide seconds per day by seconds per hour using floating-point division.

```python
seconds_per_day / seconds_per_hour
```

## Section 3.6
Divide using integer division. Does it match the previous answer?

```python
seconds_per_day // seconds_per_hour
```
